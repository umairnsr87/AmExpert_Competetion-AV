```
import pandas as pd
import numpy as np 
```

```

campaign_data=pd.read_csv("/content/drive/My Drive/buddy/amexpert/campaign_data.csv")
train_df=pd.read_csv("/content/drive/My Drive/buddy/amexpert/train.csv")
coupon_item_mapping=pd.read_csv("/content/drive/My Drive/buddy/amexpert/coupon_item_mapping.csv")
customer_demo=pd.read_csv("/content/drive/My Drive/buddy/amexpert/customer_demographics.csv")
transactions=pd.read_csv("/content/drive/My Drive/buddy/amexpert/customer_transaction_data.csv")
item_data=pd.read_csv("/content/drive/My Drive/buddy/amexpert/item_data.csv")
```

```
#step 2:drop the duplicates

item_data.drop_duplicates(['item_id', 'brand', 'brand_type', 'category'],inplace=True)#
transactions.drop_duplicates(['date', 'customer_id', 'item_id', 'quantity', 'selling_price',
       'other_discount', 'coupon_discount'],inplace=True)
train_df.drop_duplicates(['campaign_id', 'coupon_id', 'customer_id', 'redemption_status'],inplace=True)
```
